```python
#fazer leitura do link
dados = np.loadtxt(url, delimiter=',',usecols=np.arange(1,6,1), skiprows=1)
# dados
```

```python
diametro_laranja = dados[:5000,0]
peso_laranja = dados[:5000,1]

diametro_toranja = dados[5000:,0]
peso_toranja = dados[5000:,1]
```

```python
plt.plot(diametro_laranja, peso_laranja)

plt.show()
```

```python
plt.plot(diametro_toranja, peso_toranja)

plt.show()
```

```python
#y é a variável dependente (o que queremos prever).

#x é a variável independente (o fator que influencia y).

#a é o coeficiente angular (inclinação da reta), que indica a força e direção da relação entre x e y.

#b é o coeficiente linear (intercepto), que indica o valor de y quando x é igual a zero.

#a = coeficiente angular
#n = número de elementos
#Y = Moscow - preços
#x = datas

Y_laranja = peso_laranja
X_laranja = diametro_laranja
n_laranja = np.size(X_laranja)
```

```python
#calculando coeficiente angular
a_laranja = (n_laranja*np.sum(X_laranja*Y_laranja)-np.sum(X_laranja)*np.sum(Y_laranja))/(n_laranja*np.sum(X_laranja**2)-np.sum(X_laranja)**2)
a_laranja
```

```python
#Calculando o coeficiente linear
b_laranja = np.mean(Y_laranja) - a_laranja*np.mean(X_laranja)
b_laranja
```

```python
y_laranja = a_laranja*X_laranja+b_laranja
y_laranja
```

```python
np.linalg.norm(Y_laranja - y_laranja)
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(X_laranja,y_laranja)

#estimativa de valor
plt.plot(15,15*a_laranja+b_laranja,'*r') #*r faz o ponto na reta

# #estimativa de valores no futuro
# plt.plot(100,100*a+b,'*r') #*r faz o ponto na reta

plt.show()
```
