```python
#formula calcular coeficiente angular
#a = coeficiente angular
#n = número de elementos
#Y = Moscow - preços
#x = datas

Y = Moscow
X = datas
n = np.size(Moscow) # ou len(Moscow)
```

```python
(X**2).shape
```

```python
#y é a variável dependente (o que queremos prever).

#x é a variável independente (o fator que influencia y).

#a é o coeficiente angular (inclinação da reta), que indica a força e direção da relação entre x e y.

#b é o coeficiente linear (intercepto), que indica o valor de y quando x é igual a zero.

#calculando coeficiente angular
a = (n*np.sum(X*Y)-np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
a
```

image.png

```python
#Calculando o coeficiente linear
b = np.mean(Y) - a*np.mean(X)
b
```

```python
y = a*X+b
y
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x,y)

#estimativa de valor
plt.plot(41.5,41.5*a+b,'*r') #*r faz o ponto na reta

#estimativa de valores no futuro
plt.plot(100,100*a+b,'*r') #*r faz o ponto na reta

plt.show()
```
